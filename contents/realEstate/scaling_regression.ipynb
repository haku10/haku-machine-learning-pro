# 外れ値除去（3σ法）

```python
col = 'x6'
```

```python
# 平均
mean = df.mean()
mean
```

```python
mean[col]
```

```python
# 標準偏差(standard deviation)
sigma = df.std()
sigma
```

```python
sigma[col]
```

```python
low = mean[col] - 3 * sigma[col]
low
```

```python
high = mean[col] + 3 * sigma[col]
high
```

```python
df2 = df[(df[col] > low) & (df[col] < high)]
```

```python
len(df)
```

```python
len(df2)
```

```python
# 分布の確認
sns.distplot(df['x6']) # オリジナル
```

```python
# 分布の確認
sns.distplot(df2['x6']) # オリジナル
```

```python
cols = df.columns
cols
```

```python
_df = df
for col in cols:
    # 3σ法の上下限値を設定
    low = mean[col] -3 * sigma[col]
    high = mean[col] + 3 * sigma[col]
    # 条件での絞り込み
    _df = _df[(_df[col] > low) & (_df[col] < high)]
```

対処法

- ハズレ値は取り除く
- ハズレ値を平均もしくは中央値などで埋める
- 主成分分析等を使って潜在変数に変換したあとで3σ法を適応 <- 高度
